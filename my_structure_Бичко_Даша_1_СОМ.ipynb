```python
# №3. (через класи class). Тут в мене трохи не так, це не через class, але так я зробила, як зрозуміла.

students = []

n = int(input("Введіть кількість студентів, яку ви хочете дізнатися:"))

for i in range(n):
    print("Введіть дані для студента:")
    name = input("П.І.П.: ")
    birth = input("Дата народження: ")
    address = input("Адреса: ")
    course = int(input("Курс: "))
    group = input("Група: ")

    student = [name, birth, address, course, group]
    students.append(student)

print("Прізвища студентів 2-го курсу:")
for student in students:
    if student[3] == 2:
        print(student[0].split()[0])
```

```python
# № 4. (через словники dict)
cars = [
    {"марка": "Toyota ", "рік": 2018, "ціна": 35000},
    {"марка": "BMW X5", "рік": 2020, "ціна": 60000},
    {"марка": "Duster", "рік": 2019, "ціна": 32000},
    {"марка": "Skoda", "рік": 2017, "ціна": 28000},
    {"марка": "Audi", "рік": 2021, "ціна": 45000}
]

print("Автомобілі на складі:")
for car in cars:
      print("Марка: " + car['марка'] + ", Рік: " + str(car['рік']) + ", Ціна: " + str(car['ціна']) + " грн")

print("\nАвтомобілі з ціною менше 40000 грн:")
for car in cars:
    if car["ціна"] < 40000:
        print("Марка: " + car['марка'] + ", Рік: " + str(car['рік']))
```

```python
# № 4. (через класи class)
class Car:
    pass

car1 = Car()
car1.маркa = "Toyota"
car1.рік = 2018
car1.ціна = 35000

car2 = Car()
car2.маркa = "BMW X5"
car2.рік = 2020
car2.ціна = 60000

car3 = Car()
car3.маркa = "Duster"
car3.рік = 2019
car3.ціна = 32000

car4 = Car()
car4.маркa = "Skoda"
car4.рік = 2017
car4.ціна = 28000

car5 = Car()
car5.маркa = "Audi"
car5.рік = 2021
car5.ціна = 45000

cars = [car1, car2, car3, car4, car5]

print("Автомобілі на складі:")
for car in cars:
    print("Марка: " + car.маркa + ", Рік: " + str(car.рік) + ", Ціна: " + str(car.ціна) + " грн")

print("Автомобілі з ціною менше 40000 грн:")
for car in cars:
    if car.ціна < 40000:
        print("Марка: " + car.маркa + ", Рік: " + str(car.рік))
```
